### Задача 3.
### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.info()
```

```python
df['price_log'] = df['price'].apply(lambda x: np.log(x))
```

Для оценки взаимосвязей между признаками построим матрицу корреляций. Выведем матрицу корреляций и тепловую карту, чтобы лучше понимать, какие признаки имеют самую большую связь с ценой недвижимости.

```python
corr_matrix = df.corr()
```

```python
plt.figure(figsize=(21,21))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
```

По графику можно сделать следующие выводы: 

- Сильная положительная корреляция между ценой и признаками `sqft_living`, `grade`, `sqft_above`, `bathrooms`, `view` и `waterfront`.
- Слабая положительная корреляция между ценой и признаками `sqft_basement`, `bedrooms`, `floors`, `lat`.
- Сильная отрицательная корреляция между ценой и признаком `yr_built`. 

##### Зависимость цены от площади жилья

```python
plt.figure(figsize = (10, 4))
plt.scatter(df['sqft_living'], df['price_log'])
plt.xlabel('sqft_living', fontsize = 15)
plt.ylabel('price', fontsize = 15)
plt.title('Price-Sqft_living plot', fontsize = 18);
```

На графике мы видим сильную положительную корреляцию между ценой и площадью недвижимости. Больший размер площади, означает более высокую цену. 

#### Зависимость цены от качества жилья

```python
plt.figure(figsize = (10,7))
sns.boxplot(x ='grade', y ='price_log', data = df)
plt.xlabel('Grade', fontsize = 15)
plt.ylabel('Price', fontsize = 15)
plt.title('Price-Grade plot', fontsize = 18)
```

На графике мы видим сильную положительную корреляцию между ценой и качеством недвижимости. Более высокий уровень качества, означает более высокую цену. 

#### Зависимость цены от наличия береговой линии

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='waterfront', y='price_log', data=df)
plt.xlabel('Waterfront', fontsize=15)
plt.ylabel('Price', fontsize=15)
plt.title('Price-Waterfront plot', fontsize=18)
plt.xticks([0,1], ['No', 'Yes']);
```

Вид дома на набережную имеет более высокое значение цены, в сравнении с недвижимостью без доступа к берегу.

#### Зависимость цены от вида из дома

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='view', y='price_log', data=df)
plt.xlabel('View', fontsize=15)
plt.ylabel('Price', fontsize=15)
plt.title('Price-View plot', fontsize=18);
```

Цены на дома с хорошим видом более высокие.

#### Зависимость цены от количества спален и ванных комнат

```python
fig, axs = plt.subplots(ncols=2, figsize=(14,7))
sns.scatterplot(x='bedrooms', y='price_log', data=df, ax=axs[0])
axs[0].set_xlabel('Bedrooms', fontsize=15)
axs[0].set_ylabel('Price', fontsize=15)
axs[0].set_title('Price-Bedrooms plot', fontsize=18)

sns.scatterplot(x='bathrooms', y='price_log', data=df, ax=axs[1])
axs[1].set_xlabel('Bathrooms', fontsize=15)
axs[1].set_ylabel('Price', fontsize=15)
axs[1].set_title('Price-Bathrooms plot', fontsize=18)
```

На графиках мы видим, что цена недвижимости увеличивается с увеличением количества спален и ванных комнат

#### Зависимость цены от расположения жилья

```python
plt.figure(figsize=(10,7))
plt.scatter(df['lat'], df['price_log'])
plt.xlabel('Latitude', fontsize=15)
plt.ylabel('Price', fontsize=15)
plt.title('Price-Latitude plot', fontsize=18);
```

На графике мы видим, что цена недвижимости не зависит от широты расположения, но наиболее высокие цен находятся в районах с координатами 47,6-47,8.

#### Данный анализ показал, что:

##### -Наиболее важными характеристиками недвижимости для определения ее цены являются площадь жилья, качество жилья, наличие береговой линии вблизи и вид из дома.
##### - Недвижимость с большим количеством спален или ванных комнат также имеет более высокую цену.
###### - Расположение на севере или юге не оказывает существенного влияния на цену недвижимости, но наиболее высокие цены находятся в районах с координатами 47,6-47,8.
